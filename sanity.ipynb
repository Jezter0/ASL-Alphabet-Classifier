```python
import cv2
import numpy as np
import tensorflow as tf

model = tf.keras.models.load_model("asl_best_model.h5")

CLASS_NAMES = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")

# Load image
img = cv2.imread("/home/eatlunches/Documents/Codes/ComputerVision/AOL/image.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Preprocess
img_resized = cv2.resize(img, (224, 224))
img_resized = tf.keras.applications.efficientnet.preprocess_input(img_resized)
img_resized = np.expand_dims(img_resized, axis=0)

# Predict
pred = model.predict(img_resized)
probs = tf.nn.softmax(pred[0]).numpy()

print("Raw prediction:", pred)
print(dict(zip(CLASS_NAMES, probs)))

index = np.argmax(probs)
cls = CLASS_NAMES[index]

print("Prediction:", cls)
```
